```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris # iris dataset

def species_to_int(species: str) -> int:
    if species == "Iris-versicolor": return 0
    elif species == "Iris-setosa": return 1
    elif species == "Iris-virginica": return 2

def species_to_string(species: int) -> str:
    if species == 0: return "Iris-versicolor"
    elif species == 1: return "Iris-setosa"
    elif species == 2: return "Iris-virginica"

iris = load_iris() 

iris_x = iris.data 
iris_y = iris.target 
```

```python
# train/test set
x_train, x_test, y_train, y_test = train_test_split(iris_x, iris_y, test_size = 0.3)  

print("test set : ", len(y_train), "train + ", len(y_test), "test") 


# 회귀모델 생성 
model = LinearRegression() 
model.fit(X = x_train, y = y_train) 


# 모델 예측치 
y_pred = model.predict(X=x_test) # 예측치 
y_true = y_test # 정답 


# 모델 평가 : mse, corr 
mse = mean_squared_error(y_true, y_pred) 

import pandas as pd 
df = pd.DataFrame({'y_pred' : y_pred, 'y_true' : y_true}) 
#df["y_true"] = df["y_true"].apply(species_to_string)
print("\n선형 회귀 예측 결과(0 : versicolor, 1 : setosa, 2: virginica):")
print(df.to_string())
cor = df['y_pred'].corr(df['y_true']) 
print("\nACCURACY SCORE: ", cor)
```
